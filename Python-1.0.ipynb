1. **Variables:** In Python, a variable is a symbolic name that represents a value stored in the computer's memory. Variables are used to store the data that we want to manipulate in our programs.
   - With Python, we create a variable and assign a value to it simultaneously with the following syntax:
                     
                      `a_variable = a_value`
                        /                 \
                       /                   \
              Variable name          assigned value to variable 
                    
2. **Lists:** A list is a very particular type of variable because it can contain several values and its creation is done with a specific syntax:
   - a_list = [2, 3.0, "Hello"]
   - The square brackets [ and ] determine the start and the end of the list.
   - The elements of the list are separated by commas.
   - Slicing is a special kind of indexation that is used to build new lists out of existing ones.



```python
#indexing
list = [1,2,4,5,6,9]

print(list[3])
print(list[-2])
```

```python
# slicing 
my_list = [1, 5, "Hello", -1.4, "how", 103, "are", "you"]

# Retrieve the FIRST 4 elements of my_list
first_elements = my_list[0:4]

# Display these elements
print(first_elements)
```

```python

a_list = [0, 1, 2, 3, 4, 5]

# Retrieve the first 4 elements of a_list
first_4_elements = a_list[:4]
print("First 4 elements:", first_4_elements)

# Retrieve the last 3 elements of a_list
last_3_elements = a_list[-3:]
print("Last 3 elements:", last_3_elements)
```

```python
my_list1 = [1, 5, "Hello", -1.4, "how", 103, "are", "you"]

# Elements to be removed
elements_to_remove = ["Hello", "how", "are", "you"]

# Remove the specified elements from my_list1
for element in elements_to_remove:
    if element in my_list1:
        my_list1.remove(element)

# Display the modified my_list1
print("Updated my_list1:", my_list1)
```

```python
my_list = [1, 5, "Hello", -1.4, "how", 103, "are", "you"]

# Remove all numbers from my_list using the pop method
index = 0
while index < len(my_list):
    if isinstance(my_list[index], (int, float)):
        my_list.pop(index)
    else:
        index += 1

# Elements to insert
elements_to_insert = ["Bonjour", "comment", "ça", "va"]

# Indices to insert elements at
indices_to_insert = [1, 3, 5, 7]

# Insert elements into my_list at the specified indices
for element, index in zip(elements_to_insert, indices_to_insert):
    my_list.insert(index, element)

# Display the modified my_list
print("Updated my_list:", my_list)
```

```python
import pandas as pd
import numpy as np

# Generate example dataset for Sea Surface Temperature (SST)
np.random.seed(42)

# Dataset 1 (in Fahrenheit)
data_station_a = {
    'Date': pd.date_range('2023-01-01', '2023-01-10'),
    'SST_Fahrenheit': np.random.uniform(28, 32, size=10)
}

df_station_a = pd.DataFrame(data_station_a)

# Dataset 2 (in Celsius)
data_station_b = {
    'Date': pd.date_range('2023-01-01', '2023-01-10'),
    'SST_Celsius': np.random.uniform(10, 20, size=10)
}

df_station_b = pd.DataFrame(data_station_b)

# Display original datasets
print("Original Dataset 1 (Station A):")
print(df_station_a)

print("\nOriginal Dataset 2 (Station B):")
print(df_station_b)

# Harmonize data by converting Fahrenheit to Celsius for Dataset 1
df_station_a['SST_Celsius'] = (df_station_a['SST_Fahrenheit'] - 32) * 5/9

# Now, both datasets are represented in Celsius for harmonization
harmonized_data = pd.merge(df_station_a, df_station_b, on='Date')

# Display harmonized dataset
print("\nHarmonized Dataset:")
harmonized_data
```
